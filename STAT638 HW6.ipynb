# Question 7.2 

$p(y_1,..,y_n | \theta, \Sigma) \propto |\Sigma|^{-n/2} exp \left( -\frac{1}{2} \sum\limits_{i=1}^{n} (y_i - \theta)^T \Sigma^{-1}(y_i - \theta) \right)$

Using $\Psi = \Sigma^{-1}$

$\Rightarrow p(y_1,..,y_n | \theta, \Psi) \propto |\Psi|^{n/2} exp \left( -\frac{1}{2} \sum\limits_{i=1}^{n} (y_i - \theta)^T \Psi(y_i - \theta) \right) $

With $ l(\theta, \Psi | Y) = \log{p(y_1,..,y_n | \theta, \Psi)}$

$l(\theta, \Psi | Y) \propto \frac{n}{2} log{|\Psi|} - \frac{1}{2} \sum\limits_{i=1}^{n} (y_i - \theta)^T \Psi (y_i - \theta) $

Using the trace trick $a^T Ba = tr(aa^TB)$,

$l(\theta, \Psi | Y) \propto \frac{n}{2} log{|\Psi|} - \frac{1}{2} tr \left( \sum\limits_{i=1}^{n} (y_i - \theta)(y_i - \theta)^T \Psi  \right) $

$log{p_U(\theta,\Psi)} = \frac{1}{n}l(\theta, \Psi | Y)$

$log{ p_U(\theta,\Psi)} \propto \frac{1}{2} log{|\Psi|} - \frac{1}{2n} tr \left( \sum\limits_{i=1}^{n} (y_i - \theta)(y_i - \theta)^T \Psi  \right)$

$\quad \quad \quad \quad \quad = \frac{1}{2} log{|\Psi|} - \frac{1}{2n} tr \left( \sum\limits_{i=1}^{n} (y_i - \theta)(y_i - \theta)^T \Psi  \right)$

Using the standard trick of adding and subtracting $\bar{y}$

$\quad \quad \quad \quad \quad = \frac{1}{2} log{|\Psi|} - \frac{1}{2n} tr \left( \sum\limits_{i=1}^{n} (y_i - \bar{y})(y_i - \bar{y})^T \Psi   + n (\bar{y} - \theta)(\bar{y} - \theta)^T  \Psi\right)$

$\quad \quad \quad \quad \quad = \frac{1}{2} log{|\Psi|} - \frac{1}{2n} \sum\limits_{i=1}^{n} (y_i - \bar{y})^T\Psi(y_i - \bar{y})   -\frac{1}{2} (\bar{y} - \theta)^T \Psi(\bar{y} - \theta) $


$\Rightarrow p_U(\theta, \Psi) \propto |\Psi|^{1/2} \exp\left( -\frac{1}{2n} \sum\limits_{i=1}^{n} (y_i - \bar{y})^T\Psi(y_i - \bar{y}) \right) \exp \left( -\frac{1}{2} (\theta - \bar{y})^T\Psi(\theta - \bar{y}) \right) $

Writing this as, 

$p_U(\theta | \Psi) \propto |\Psi|^{1/2} \exp \left( -\frac{1}{2} (\theta - \bar{y})^T\Psi(\theta - \bar{y}) \right) $ which is $MVN(\bar{y}, \Psi^{-1})$ and 

Using $S = \frac{1}{n}\sum\limits_{i=1}^{n} (y_i - \theta)(y_i - \theta)^T  $,


$p_U(\Psi) \propto |\Psi|^{(p+1-p-1)/2} \exp\left( -\frac{1}{2n} \sum\limits_{i=1}^{n} (y_i - \bar{y})^T\Psi(y_i - \bar{y}) \right)  $ which is $Wishart(p+1,S)$ 

$\Sigma$ follows $inverseWishart(p+1,S^{-1})$

$p_U(\theta, \Sigma | y_1, ...., y_n) \propto  |\Sigma|^{-1/2} \exp\left( -\frac{1}{2} (\theta - \bar{y})^T\Sigma(\theta - \bar{y}) \right) |\Sigma|^{-(2p+2)/2} \exp{-\frac{1}{2} tr(S\Sigma^{-1}) }   $ 

$\quad \quad \quad \quad \quad \times |\Sigma|^{-(n)/2}  \exp{-\frac{1}{2} \sum\limits_{i=1}^{n} (y_i -\bar{y})^T \Sigma^{-1} (y_i -\bar{y}) }  \exp{-\frac{n}{2}  (\theta -\bar{y})^T\Sigma^{-1} (\theta -\bar{y}) }  $

$ \exp{-\frac{n}{2}  (\theta -\bar{y})^T\Sigma^{-1} (\theta -\bar{y}) } \exp\left( -\frac{1}{2} (\theta - \bar{y})^T\Sigma(\theta - \bar{y}) \right) $

$ \quad \quad \quad \quad \quad = \exp{-\frac{n+1}{2}  (\theta -\bar{y})^T\Sigma^{-1} (\theta -\bar{y}) }  $

$ \exp{-\frac{1}{2} \sum\limits_{i=1}^{n} (y_i -\bar{y})^T \Sigma^{-1} (y_i -\bar{y}) }  =  \exp{-\frac{1}{2} tr( nS \Sigma^{-1})  } $

Using which, 

$p_U(\theta, \Sigma | y_1, ...., y_n) \propto |\Sigma|^{-1/2} \exp\left( -\frac{n+1}{2} (\theta -\bar{y})^T\Sigma^{-1} (\theta -\bar{y}) \right)  |\Sigma|^{-(2p+n+2)/2} exp \left( -\frac{1}{2} tr(n+1) S\Sigma^{-1} \right) $

whicc is multivariate normal inversewishart distribution. Since unit information prior can be considered a prior, this can be considered as a posterior.
