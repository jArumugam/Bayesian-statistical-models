# Question 5.1 

Given {$\mu_0, \sigma_0, k_0, \nu_0$} and n data points $y$ =  {$y_1, y_2, ..., y_n$}

$\{1/\sigma^2\} \sim gamma(\nu_0, \nu_0 \sigma_0^2/2)$

$\{\theta| \sigma^2\} \sim normal(\mu_0, \sigma_0^2/k_0)$

$\bar{y} = mean(y)$, and $\sigma^2 = variance(y)$

$k_n = k_0 + n$

$\nu_n = \nu_0 + n$ 

$\mu_n = 1/k_n (k_0 \mu_0 + n \bar{y}) $ 

$\sigma^2_n = 1/\nu_n (\nu_0 \sigma^2_0 + (n-1)\sigma^2 + \frac{k_0 n}{k_n} (\bar{y} - \mu_0)^2 )$ 

$\{1/\sigma^2 | y\} \sim gamma(\nu_n, \nu_n \sigma_n^2/2)$ 

$\{\theta | \sigma^2, y\} \sim normal(\mu_n, \sigma^2/k_n)$

```python
with open('school1.dat') as f:
    yA = [[float(x) for x in line.split()] for line in f]
yA = reduce(operator.add, yA)    

with open('school2.dat') as f:
    yB = [[float(x) for x in line.split()] for line in f]
yB = reduce(operator.add, yB)    

with open('school3.dat') as f:
    yC = [[float(x) for x in line.split()] for line in f]
yC = reduce(operator.add, yC) 

lyA = len(yA)
yAmean = sum(yA)/lyA
yAvar = sum( (yAmean - value) ** 2 for value in yA ) / lyA

lyB = len(yB)
yBmean = sum(yB)/lyB
yBvar = sum( (yBmean - value) ** 2 for value in yB ) / lyB

lyC = len(yC)
yCmean = sum(yC)/lyC
yCvar = sum( (yCmean - value) ** 2 for value in yC ) / lyC

print "Sample A mean and variance :", yAmean, yAvar
print "Sample B mean and variance :", yBmean, yBvar
print "Sample C mean and variance :", yCmean, yCvar
```

```python
mu0, var0, k0, nu0 = 5., 4., 1., 2.

knA = k0 + lyA
nuNA = nu0 + lyA
muNA = (k0*mu0 + lyA*yAmean)/knA
varNA = 1/nuNA*( nu0*var0 + (lyA-1)*yAvar + k0/knA*lyA*(yAmean - mu0)**2)

knB = k0 + lyB
nuNB = nu0 + lyB
muNB = (k0*mu0 + lyB*yBmean)/knB
varNB = 1/nuNB*( nu0*var0 + (lyB-1)*yBvar + k0/knB*lyB*(yBmean - mu0)**2)

knC = k0 + lyC
nuNC = nu0 + lyC
muNC = (k0*mu0 + lyC*yCmean)/knC
varNC = 1/nuNC*( nu0*var0 + (lyC-1)*yCvar + k0/knB*lyC*(yCmean - mu0)**2)

print knA, nuNA, muNA, varNA
print knB, nuNB, muNB, varNB
print knC, nuNC, muNC, varNC
```

```python
aA, bA = nuNA/2., nuNA*varNA/2.
aB, bB = nuNB/2., nuNB*varNB/2.
aC, bC = nuNC/2., nuNC*varNC/2.

nSample = 50000

oneByVarA = st.gamma.rvs(aA, size=nSample, scale = 1./bA)
oneByVarB = st.gamma.rvs(aB, size=nSample, scale = 1./bB)
oneByVarC = st.gamma.rvs(aC, size=nSample, scale = 1./bC)

sVarA = [ 1./i for i in oneByVarA]
sVarB = [ 1./i for i in oneByVarB]
sVarC = [ 1./i for i in oneByVarC]

sThA = [0. for i in range(0,nSample)]
sThB = [0. for i in range(0,nSample)]
sThC = [0. for i in range(0,nSample)]

sYA = [0. for i in range(0,nSample)]
sYB= [0. for i in range(0,nSample)]
sYC= [0. for i in range(0,nSample)]

for i in range(0,nSample):
    sThA[i] = np.random.normal(muNA, scale = math.sqrt(sVarA[i]/knA))
    sThB[i] = np.random.normal(muNB, scale = math.sqrt(sVarB[i]/knB))
    sThC[i] = np.random.normal(muNC, scale = math.sqrt(sVarC[i]/knC))
    
    sYA[i] = np.random.normal(sThA[i], scale = math.sqrt(sVarA[i]/knA))
    sYB[i] = np.random.normal(sThB[i], scale = math.sqrt(sVarB[i]/knB))
    sYC[i] = np.random.normal(sThC[i], scale = math.sqrt(sVarC[i]/knC))
```

```python
msThA = sum(sThA)/len(sThA)
varsThA = sum(sVarA) / len(sVarA)

msThB = sum(sThB)/len(sThB)
varsThB = sum(sVarB) / len(sVarB)

msThC = sum(sThC)/len(sThC)
varsThC = sum(sVarC) / len(sVarC)

print " "
print "School 1 mean of theta and variance : ", msThA, varsThA
print "School 1 mean of theta and variance : ", msThB, varsThB
print "School 1 mean of theta and variance : ", msThC, varsThC
print " "
```

```python
k = [1.,2.,4.,8.,16.,32.]

prob = [0. for i in range(0,6)]

for j in range(0,6):
    mu0, var0 = 75, 100.
    k0, nu0 = k[j], k[j]
    
    y1mean = 75.2
    y1var = 7.3**2
    y2mean = 77.5
    y2var = 8.1 ** 2
    ly1 = 16.
    ly2 = 16.
    
    kn1 = k0 + ly1
    nuN1 = nu0 + ly1
    muN1 = (k0*mu0 + ly1*y1mean)/kn1
    varN1 = 1/nuN1*( nu0*var0 + (ly1-1.)*y1var + k0/kn1*ly1*(y1mean - mu0)**2)
    
    kn2 = k0 + ly2
    nuN2 = nu0 + ly2
    muN2 = (k0*mu0 + ly2*y2mean)/kn2
    varN2 = 1/nuN2*( nu0*var0 + (ly2-1.)*y2var + k0/kn2*ly2*(y2mean - mu0)**2)
    
    a1, b1 = nuN1/2., nuN1*varN1/2.
    a2, b2 = nuN2/2., nuN2*varN2/2.
    
    oneByVar1 = st.gamma.rvs(a1, size=nSample, scale = 1./b1)
    oneByVar2 = st.gamma.rvs(a2, size=nSample, scale = 1./b2)
    
    sVar1 = [ 1./i for i in oneByVar1]
    sVar2 = [ 1./i for i in oneByVar2]
    
    sTh1 = [0. for i in range(0,nSample)]
    sTh2 = [0. for i in range(0,nSample)]
    
    nCount = 0.
    for i in range(0,nSample):
        sTh1[i] = np.random.normal(muN1, scale = math.sqrt(sVar1[i]/kn1))
        sTh2[i] = np.random.normal(muN2, scale = math.sqrt(sVar2[i]/kn2))
        if sTh1[i] < sTh2[i]:
            nCount += 1
    
    prob[j] = nCount/nSample
```

```python
plt.figure()
plt.plot(k, prob, '-o')
plt.show
```

With $k_0 = \nu_0$, the probability is higer for small values of  $k_0$, prior information is less and data favors $\theta_B > \theta_A$. As the prior gets stronger (large values of $ k_0$), still  $\theta_B > \theta_A$ is favored. 

# Question 5.5

$p(y | \theta, \sigma) = \frac{1}{{2 \pi}^{1/2} \sigma} e^{ \left( - \frac{1}{2\sigma^2} (y - \mu)^2 \right)} $ 

With $\psi = 1/\sigma^2$

$p(y | \theta, \pi) = \frac{ {\psi}^{1/2}}{ {2 \pi}^{1/2}} e^{ \left( -\frac{\psi}{2} (y - \mu)^2 \right)} $ 

$ \log{p(y | \theta, \pi)} = -\frac{1}{2} \log{2 \pi} + \frac{1}{2} \log{\psi} 
- \psi/2 ( y - \theta )^2$ 

$ l(\theta, \pi | y_1, y_2,..., y_n) = \sum \log{p(y | \theta, \pi)} = -\frac{n}{2} \log{2 \pi} + \frac{n}{2} \log{\psi} 
- \sum \frac{\psi}{2} ( y - \theta )^2$ 

Using the hint, 

$l(\theta, \pi | {Y}) =  -\frac{n}{2} \log{2 \pi} + \frac{n}{2} \log{\psi} - \frac{\psi}{2} \sum ( y- \bar{y} )^2 - n \frac{\psi}{2} ( \bar{y} - \theta )^2$ 

Using $\log{p_U(\theta, \psi)} = l(\theta, \pi | {Y})/n $,  and $ \sigma^2 = 1/n \sum ( y- \bar{y} )^2 $

$ \log{p_U(\theta, \psi)} = C + \frac{1}{2} \log{\psi} - \psi \sigma^2 -  \psi ( \bar{y} - \theta )^2 $ 

$\Rightarrow p_U(\theta, \psi) \propto \psi^{1/2} e^{-\frac{\psi}{2} \sigma^2} e^{-\frac{\psi}{2}(\bar{y} - \theta)^2} $ 

$ \Rightarrow p_U(\psi) \sim gamma(3/2, \sigma^2/2) $ and 
$p_U(\theta | \psi) \sim normal( \bar{y}, 1/\psi) $

$p(y_1, y_2, ..., y_n | \theta, \psi) = p_U(\theta, \psi) p(\theta, \sigma | y_1, y_2, ..., y_n) $

$p(y_1, y_2, ..., y_n | \theta, \psi) \propto  \psi^{1/2} e^{-\frac{\psi}{2} \sigma^2} e^{-\frac{\psi}{2}(\bar{y} - \theta)^2}  
{\psi}^{n/2} e^{ \left( -\frac{\psi}{2} \sum (y - \mu)^2 \right)} $

This looks like some normal-gamma distribution. 

Since the unit information prior can be roughly viewed as the prior information, this can be roughly viewed as a posterior distribution. But the effect of data is considered in prior and hence we its use should be cautious.
