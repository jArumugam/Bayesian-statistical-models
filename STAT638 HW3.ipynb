# Question 3.14 

Log likelihood is $l(\theta|y) = \sum log\  p(y_i | \theta)$

The sufficient statistic is $\sum y_i = y$

## 3.14 (A)
Likelihood, $ p(y | \theta) = \binom{n}{y} \theta^{y} (1 - \theta)^{n-y} $

Log likelihood, $ l(\theta | y) = C_y + y log \theta + (n-y) log (1-\theta) $

$ \frac{\partial}{\partial \theta} l(\theta | y) = \frac{y}{\theta} - \frac{n-y}{1-\theta}$

$ \frac{\partial}{\partial \theta} l(\theta | y) =  0 \Rightarrow y - y \theta = n \theta - y \theta $ 

$\Rightarrow \hat{\theta} = \frac{y}{n}$ is the MLE.

$ \frac{\partial}{\partial \theta} l(\theta | y) = -\frac{y}{\theta^2} - \frac{n-y}{(1-\theta)^2}$ 

$\Rightarrow J(\theta) = \frac{y}{\theta} + \frac{n+y}{(1-\theta)^2} $ 

$ J(\hat{\theta}) = n + \frac{n^2}{n-y} $ 

$ J(\hat{\theta}) = 1 + \frac{n}{n-y} $

## 3.14 (B)

$\log{p_U(\theta)} = \frac{l(\theta | y)}{n} + c$ 

$\log{p_U(\theta)} = \frac{y}{n} \log{\theta} +  (1 - \frac{y}{n}) \log{1-\theta} + C_1$

$\log{p_U(\theta)} = \log{\theta} +  \frac{y}{n} \log{\frac{\theta}{1-\theta}} + \log{C_1}$ 

$\log{p_U(\theta)} = \log{ C_1 (1-\theta) (\frac{\theta}{1-\theta})^{\frac{y}{n}} }$ 

$ \Rightarrow {p_U(\theta)} = { C_1 (1-\theta) (\frac{\theta}{1-\theta})^{\frac{y}{n}} } 
= C_1 \theta^{\frac{y}{n}} (1 - \theta)^{\frac{n-y}{n}}$ 

$ \frac{\partial}{\partial \theta} \log{p_U(\theta)} = \frac{y}{n\theta} - (1 - \frac{y}{n})\frac{1}{1-\theta}$

$ \frac{\partial^2}{\partial \theta^2} \log{p_U(\theta)} = -\frac{y}{n\theta^2} - (1 - \frac{y}{n})\frac{1}{(1-\theta)^2}$ 

$ J(\theta) = \frac{y}{n\theta^2} + (1 - \frac{y}{n})\frac{1}{(1-\theta)^2}$

## 3.14 (C)

$ p_U(\theta) p(y_1, y_2,..., y_n | \theta) \propto \theta^{\frac{y}{n}} (1 - \theta)^{\frac{n-y}{n}} \theta^{{y}} (1 - \theta)^{{n-y}} $ 

$p_U(\theta) p(y_1, y_2,..., y_n | \theta) \propto \theta^{y(\frac{1}{n} + 1)} (1 - \theta)^{n -y + 1 -\frac{y}{n}} $ 

Thios is $beta \left(\frac{y}{n} + y + 1, n -y + 2 -\frac{y}{n} \right)$ distribution. Since the question mentions *the unit information prior can be roughly viewed as the prior information*, this can be roughly viewed as a posterior distribution. But the effect of data is considered in prior and hence we its use should be cautious. 

## 3.14 (d)

Poisson likelihood $ L(\theta | y) = \prod\limits_{i=1}^{n} \frac{e^{-\theta} \theta^{y_i}}{y_i !}$

$l(\theta | y) = \log{L(\theta | y) }  = -n\theta + y \log{\theta} - C_y $ 

$ \frac{\partial }{\partial \theta} l(\theta | y) = -n + \frac{y}{\theta} = 0 \Rightarrow \hat{\theta} = \frac{y}{n}$

$ \frac{\partial^2 }{\partial \theta^2} l(\theta | y) = -\frac{y}{\theta^2} \Rightarrow J(\theta) = \frac{y}{\theta^2}$

$\Rightarrow J(\hat{\theta})/n = \frac{n}{y}$

$\log{p_U(\theta)} = \frac{l(\theta | y)}{n} + c = -\theta + \frac{y}{n} \log{\theta} + C_1$

$\Rightarrow p_U(\theta) \propto e^{-\theta} \theta^{y/n}$ which is a gamma distribution.

$ \frac{\partial}{\partial \theta} \log{p_U(\theta)} = -1 + \frac{y}{n\theta}$

$ \frac{\partial^2}{\partial \theta^2} \log{p_U(\theta)} = -\frac{y}{n\theta^2}$ 

$ J(\theta) = \frac{y}{n\theta^2} \Rightarrow J(\hat{\theta}) = \frac{n}{y} $

$ p_U(\theta) p(y_1, y_2,..., y_n | \theta) \propto e^{-\theta} \theta^{y/n} e^{-n\theta} \theta^{y} 
= e^{-(1+n)\theta} \theta^{y(1/n + 1)} $ 

This is a gamma distribution $gamma\left( (1+n) \frac{y}{n} + 1, 1+n \right)$. 

# Question 4.1

Prior $P(\theta_1) = 1$, $\theta$ is a continuous variable

Likelihood for a given $\theta_{1}$,
$Pr(\sum Y_{i} = 57 | \theta_{1}) = \binom {100} {57} \theta_{1}^{y} (1 - \theta_{1})^{100-57} $

Posterior has a similar form since the prior is uniform
$Pr(\theta_1 | \sum Y_{i} = 57 ) \propto \binom {100} {57} \theta_{1}^{y} (1 - \theta_{1})^{100-57} $ 

$Pr(\theta_1 | \sum Y_{i} = 57 ) \sim dbeta(58,44)$ 

Similarly posterior for $\theta_2$

$Pr(\theta_2 | \sum Y_{i} = 30 ) \propto \binom {50} {30} \theta_{2}^{y} (1 - \theta_{2})^{50-30} $ 

$Pr(\theta_2 | \sum Y_{i} = 30 ) \sim dbeta(31,21)$ 

```python
a1, b1 = 58. , 44.
a2, b2 = 31. , 21.

theta141 = st.beta.rvs(a1, b1, size=5000)
theta241 = st.beta.rvs(a2, b2, size=5000) 
```

```python
nCount= 0; 
for i in range (0,5000):
    if theta141[i] < theta241[i]:
        nCount = nCount + 1
        
print nCount/5000.
```

$Pr(\theta_1 < \theta_2 | data\ and\ prior ) \approx 0.6312 $

# Question 4.2 

$y_{A} = (12, 9, 12, 14, 13, 13, 15, 8, 15, 6);$

$y_{B} = (11, 11, 10, 9, 9, 8, 7, 10, 6, 8, 8, 9, 7);$ 

$ y_{A} \sim dpoisson(y, \theta) $ 

$Pr(y_{i} | \theta_{}) = \frac{ \theta^{y_{i}} e^{-\theta} }{y_{i} !} $ 

$dgamma(\theta, a, b) = \frac{b^a}{\Gamma(a)} \theta^{a-1} e^{-b \theta} $ 

Prior: $\theta_{A} \sim dgamma(120, 10)$ 

Likelihood: $Pr(Y | \theta) = \prod\limits_{i = 1}^{n} p(y_{i} | \theta) = \frac{\theta^{ \sum y_{i}} e^{-n{\theta}}}
{\prod\limits_{i = 1}^{n} y_{i} ! }$


Posterior: $Pr(\theta | Y) \propto Pr(Y | \theta) Pr(\theta) $

$Pr(\theta | Y) \propto \theta^{\sum y_{i} + a - 1} e^{\theta (n + b)} $

$\sum y_i = 117$. 

$\sum y_{i} + a - 1 = 236$

$n + b = 10 + 10 = 20$

$Pr(\theta_A | Y_A) \sim dgamma(237,20)$ 

$\theta_{B} \sim dgamma(12 n_0, n_0)$ where $n \in \{1,2,...,50\}$ 

Posterior: $Pr(\theta | Y_B) \propto Pr(Y_B | \theta_B) Pr(\theta_B) $

$Pr(\theta_B | Y_B) \propto \theta_B^{\sum y_{i} + 12 n_0 - 1} e^{\theta (n + n_0)} $ 

$Pr(\theta_B | Y_B) \sim dgamma(113+12n_0,13+n_0)$ 


To summarise, $Pr(\theta_B | Y_B) \sim dgamma(125,14)$ 
and 
$Pr(\theta_A | Y_A) \sim dgamma(237,20)$ 

```python
a1, b1 = 237. , 20.
a2, b2 = 125. , 14.

thetaa42a = st.gamma.rvs(a1, size=5000,  scale = 1.0/b1)
thetab42a = st.gamma.rvs(a2, size=5000, scale = 1.0/b2) 

nCount= 0; 
for i in range (0,5000):
    if thetab42a[i] < thetaa42a[i]:
        nCount = nCount + 1
        
print nCount/5000.
```

## 4.2 (C)

From the book, when prior $\theta \sim gamma(a,b)$, and likelihood $y_i | \theta \sim poisson(y_i, \theta)$, 

then the posterior $\theta | y_1, y_2,...,y_n \sim gamma(a + \sum\limits_{i=1}^{n} y_i, b + n ) $

and the postyerior predictive distribution is **negative binomial distribution** 

given by $p(\tilde{y} | y_1,...,y_n) = \frac{\Gamma(a + \sum\limits_{i=1}^{n} y_i + \tilde{y})}{\Gamma(\tilde{y}+1) \Gamma(a + \sum\limits_{i=1}^{n} y_i)} 
\left( \frac{b+n}{b+n+1} \right)^{a + \sum\limits_{i=1}^{n} y_i } 
\left( \frac{1}{b+n+1}\right)^{\tilde{y}}$ 

i.e., $p(\tilde{y} | Y) = nbinom \left( a + \sum\limits_{i=1}^{n} y_i, b+n \right) $ 

$p(\tilde{Y_A} | Y_A) = nbinom( 120 + 117, 10 + 10) = nbinom( 237, 20)$ 

$p(\tilde{Y_B} | Y_B) = nbinom (113+12n_0,13+n_0)$ 

```python
a1, b1 = 237. , 20.
a2, b2 = 125. , 14.

Ya42a = st.nbinom.rvs(a1, b1/(b1+1.), size=5000)
Yb42a = st.nbinom.rvs(a2, b2/(b2+1.), size=5000) 

nCount= 0; 
for i in range (0,5000):
    if Yb42a[i] < Ya42a[i]:
        nCount = nCount + 1
        
print nCount/5000.
```

$ Pr(\tilde{y}_B < \tilde{y}_A | y_A, y_B) \approx 0.6958 $
